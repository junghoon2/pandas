```python
import pandas as pd
```

```python
df = pd.read_csv('24.house_2020.csv')
```

```python
df
```

```python
df = df.rename(columns={'거래금액(만원)': '거래금액'})
# dict 형식 사용. { }, key, value로
```

```python
df.거래금액 = df.거래금액.str.replace(',', '').astype('int64')
```

```python
city = '용인|성남|하남|남양주|광주'
```

```python
df=df[df.시군구.str.contains(city)]
```

```python
df
```

```python
df = df [(df.거래금액 > 50000) & (df.거래금액 < 100000) & (df.건축년도 > 1999) & (df.주택유형 == '단독')]
# & 조건을 ()에 하나하나 넣어주면 된다.
```

```python
df
```

```python
df_count = df.시군구.value_counts()
#df.시군구.nunique()
```

```python
df.to_csv('13.house_price_2020.csv', encoding='utf-8-sig')
# utf-8-sig 해야 한글 깨지지 않는다.
```
